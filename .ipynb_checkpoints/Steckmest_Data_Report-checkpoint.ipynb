### Data set 2:  Recidivism across time, sex and age

The second data set is significantly larger and describes the raw numbers of people who, by categories of sex and age, were re-convicted after their release from a previous crime.
```python
# import data set from computer

file_recid = 'SSB_recidivism.xlsx'

recid_years = pd.read_excel(path+file_recid, sheetname='By_Years_Sex_Age')
recid_years = recid_years.rename(columns={'Unnamed: 1': 'Sex',
                         'Unnamed: 2': 'Age',
                         'Unnamed: 3': '2002',
                         'Unnamed: 4': '2003',
                         'Unnamed: 5': '2004',
                         'Unnamed: 6': '2005',
                         'Unnamed: 7': '2006',
                         'Unnamed: 8': '2007',
                         'Unnamed: 9': '2008',
                         'Unnamed: 10': '2009',
                         'Unnamed: 11': '2010'})
recid_years.head()
```
